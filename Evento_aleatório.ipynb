```python
import numpy as np
import matplotlib.pyplot as plt

moeda = ['cara', 'coroa']
lcaras = []
ln = []
for n in range(1,1000):
  ncaras = 0
  for i in range(0,n):
    face = np.random.choice(moeda)
    if face == 'cara':
      ncaras = ncaras + 1
  fa = ncaras/n
  lcaras.append(fa)
  ln.append(n)
plt.plot(ln, lcaras)
plt.show
```

```python
dado = [1,2,3,4,5,6]

n = 100
vfaces = np.zeros(6)

for i in range(0,n):
  face = np.random.choice(dado)
  vfaces[face-1] =  vfaces[face-1] + 1

print('Valor teórico:', 1/6)
print('Frequencias obtidas:", vfaces/n')
```
