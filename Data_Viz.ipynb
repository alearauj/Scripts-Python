```python
## Carrega dados, transforma em DataFrames e cria algumas colunas categóricas.

import matplotlib.pyplot as plt  # we still need Matplotlib for some of the packages
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine, load_iris

wine = load_wine()
wine_df = pd.DataFrame(wine.data, columns=wine.feature_names)
wine_X = wine_df.copy()
wine_y = [wine.target_names[i] for i in wine.target]
wine_df['region'] = [wine.target_names[i] for i in wine.target]

iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = [iris.target_names[i] for i in iris.target]

states_df = pd.DataFrame([
    ['WA', 0.56, False],
    ['GA', 0.31, True],
    ['IN', 0.41, True],
    ['WI', 0.77, True],
    ['CT', 0.55, False],
    ['NJ', 0.48, True],
], columns=['state', 'var1', 'var2'])
```

```python
import seaborn as sns ## great for plotting non-aggregated distributions of data

sns.boxenplot(wine_df, x='region', y='magnesium', hue='region', palette='Set1', legend=False, width=0.35)
```

```python
my_time_series_data = pd.DataFrame({
    'time': np.repeat(range(50), 10),
    'group': np.tile(['a','b','c','d','e','f','g','h','i','j'], 50),
    'observation': np.hstack([np.random.normal(10, scale=0.3, size=250),
                              np.random.normal(10, scale=0.8, size=250)])
})

sns.lineplot(my_time_series_data, x='time', y='observation', errorbar='ci')
```

```python
_ = sns.displot(data=wine_df, x='magnesium', hue='region', legend=True, kind='hist', bins=20, multiple='stack', height=4)
_ = sns.displot(data=wine_df, x='magnesium', hue='region', legend=True, kind='kde', height=4)
_ = sns.displot(data=wine_df, x='magnesium', hue='region', legend=True, kind='ecdf', height=4)
```
